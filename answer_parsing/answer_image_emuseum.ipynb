```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import os
import urllib.request

# 셀레니움 설정
options = Options()
#options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=options)

# 테스트 링크 (dataUCI.jsp 페이지)
uci_url = "http://hnkostma.org/emuseum/service/data/dataUCI.jsp?data_uci=KCIS_U038+MPNHNKSMC+KSM-XD.1881.4687-20090726.YNE00019_0"
driver.get(uci_url)
time.sleep(2)

# 🔹 JavaScript 함수 직접 실행
button = driver.find_element(By.LINK_TEXT, "상세정보")
href = button.get_attribute("href")
print("📍 버튼 href:", href)
buttons = driver.find_elements(By.CSS_SELECTOR, "a.button")

for btn in buttons:
    if "상세정보" in btn.text in btn.get_attribute("href"):
        btn.click()
        print("✔ '원문이미지' 버튼 클릭 완료")
        time.sleep(2)
        break
time.sleep(2)  # 페이지 로딩 대기

# 🔍 현재 URL 확인
print("📍 이동 후 URL:", driver.current_url)

# 🔍 이미지 추출
img_tags = driver.find_elements(By.TAG_NAME, "img")
os.makedirs("downloaded_images", exist_ok=True)

for i, img_tag in enumerate(img_tags):
    src = img_tag.get_attribute("src")
    if src and src.startswith("http"):
        filename = os.path.join("downloaded_images", f"{i+1:02d}.jpg")
        urllib.request.urlretrieve(src, filename)
        print(f"✔ 이미지 저장됨: {filename}")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# 크롬 드라이버 설정
options = Options()
options.add_argument("--headless")  # 디버깅 위해 주석처리
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 대상 페이지
url = "http://hnkostma.org/emuseum/service/data/dataUCI.jsp?data_uci=KCIS_U038+MPNHNKSMC+KSM-XD.1881.4687-20090726.YNE00019_0"
driver.get(url)
time.sleep(2)

# '원문이미지' 버튼 클릭
button = driver.find_element(By.LINK_TEXT, "상세정보")
button.click()
time.sleep(2)

# 새 창 전환
handles = driver.window_handles
if len(handles) > 1:
    driver.switch_to.window(handles[1])
    viewer_url = driver.current_url
    print("✅ 원문이미지 뷰어 URL:", viewer_url)
else:
    print("⚠ 새 창이 감지되지 않았습니다.")
    viewer_url = None

driver.quit()

# viewer_url 값을 활용할 수 있도록 반환
print("🧾 최종 URL:", viewer_url)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

# CSV 불러오기 (링크 목록)
with open("../data/sigwon/emuseum.csv", newline='') as f:
    reader = csv.reader(f)
    uci_urls = [row[0].strip() for row in reader if "dataUCI.jsp" in row[0]]

# 셀레니움 설정
options = Options()
options.add_argument("--headless")  # 디버깅 시 주석처리
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

results = []

for url in uci_urls:
    driver.get(url)
    time.sleep(2)

    try:
        button = driver.find_element(By.LINK_TEXT, "상세정보")
        button.click()
        time.sleep(2)

        handles = driver.window_handles
        if len(handles) > 1:
            driver.switch_to.window(handles[1])
            viewer_url = driver.current_url
            results.append((url, viewer_url))
            print(f"[✔] {url} → {viewer_url}")
            driver.close()
            driver.switch_to.window(handles[0])
        else:
            print(f"[✖] 새 창 없음: {url}")
            results.append((url, None))

    except Exception as e:
        print(f"[❌] 실패: {url}, 원인: {e}")
        results.append((url, None))

driver.quit()

# 결과 저장
with open("viewer_urls.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["source_url", "viewer_url"])
    writer.writerows(results)
```

```python
import pandas as pd
import re
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# 설정
input_csv = "viewer_urls.csv"       # 입력 CSV 파일 (URL 목록)
url_column = "viewer_url"           # URL이 들어있는 컬럼명
output_csv = "output.csv"    # 출력 CSV

# Selenium WebDriver 설정
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 입력 CSV 로드
df_input = pd.read_csv(input_csv)

# 결과 저장용 리스트
results = []

# URL 순회
for idx, row in df_input.iterrows():
    url = row[url_column]
    print(f"[{idx+1}/{len(df_input)}] 처리 중: {url}")
    
    try:
        driver.get(url)
        time.sleep(3)  # 렌더링 대기

        div = driver.find_element(By.ID, "sangse_res")
        raw_text = div.text.strip()

        # 개행 두 줄 이상 제거
        clean_text = re.sub(r'\n\s*\n', '\n', raw_text)

        # 제목은 첫 줄로 간주
        title = clean_text.split('\n')[0] if clean_text else ""

        results.append({
            "URL": url,
            "제목": title,
            "본문": clean_text
        })

    except Exception as e:
        print(f"❌ 오류 발생: {e}")
        results.append({
            "URL": url,
            "제목": "오류 발생",
            "본문": ""
        })

# 종료
driver.quit()

# 결과 저장
df_output = pd.DataFrame(results)
df_output.to_csv(output_csv, index=False, encoding='utf-8-sig')

print(f"✅ 전체 완료. 결과 저장: {output_csv}")
```
