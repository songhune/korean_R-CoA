```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

# CSV ë¶ˆëŸ¬ì˜¤ê¸° (ë§í¬ ëª©ë¡)
with open("../data/sigwon/emuseum.csv", newline='') as f:
    reader = csv.reader(f)
    uci_urls = [row[0].strip() for row in reader if "dataUCI.jsp" in row[0]]

# ì…€ë ˆë‹ˆì›€ ì„¤ì •
options = Options()
options.add_argument("--headless")  # ë””ë²„ê¹… ì‹œ ì£¼ì„ì²˜ë¦¬
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

results = []

for url in uci_urls:
    driver.get(url)
    time.sleep(2)

    try:
        button = driver.find_element(By.LINK_TEXT, "ìƒì„¸ì •ë³´")
        button.click()
        time.sleep(2)

        handles = driver.window_handles
        if len(handles) > 1:
            driver.switch_to.window(handles[1])
            viewer_url = driver.current_url
            results.append((url, viewer_url))
            print(f"[âœ”] {url} â†’ {viewer_url}")
            driver.close()
            driver.switch_to.window(handles[0])
        else:
            print(f"[âœ–] ìƒˆ ì°½ ì—†ìŒ: {url}")
            results.append((url, None))

    except Exception as e:
        print(f"[âŒ] ì‹¤íŒ¨: {url}, ì›ì¸: {e}")
        results.append((url, None))

driver.quit()

# ê²°ê³¼ ì €ì¥
with open("viewer_urls.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["source_url", "viewer_url"])
    writer.writerows(results)
```

```python
import pandas as pd
import re
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# ì„¤ì •
input_csv = "viewer_urls.csv"       # ì…ë ¥ CSV íŒŒì¼ (URL ëª©ë¡)
url_column = "viewer_url"           # URLì´ ë“¤ì–´ìˆëŠ” ì»¬ëŸ¼ëª…
output_csv = "output.csv"    # ì¶œë ¥ CSV

# Selenium WebDriver ì„¤ì •
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# ì…ë ¥ CSV ë¡œë“œ
df_input = pd.read_csv(input_csv)

# ê²°ê³¼ ì €ì¥ìš© ë¦¬ìŠ¤íŠ¸
results = []

# URL ìˆœíšŒ
for idx, row in df_input.iterrows():
    url = row[url_column]
    print(f"[{idx+1}/{len(df_input)}] ì²˜ë¦¬ ì¤‘: {url}")
    
    try:
        driver.get(url)
        time.sleep(3)  # ë Œë”ë§ ëŒ€ê¸°

        div = driver.find_element(By.ID, "sangse_res")
        raw_text = div.text.strip()

        # ê°œí–‰ ë‘ ì¤„ ì´ìƒ ì œê±°
        clean_text = re.sub(r'\n\s*\n', '\n', raw_text)

        # ì œëª©ì€ ì²« ì¤„ë¡œ ê°„ì£¼
        title = clean_text.split('\n')[0] if clean_text else ""

        results.append({
            "URL": url,
            "ì œëª©": title,
            "ë³¸ë¬¸": clean_text
        })

    except Exception as e:
        print(f"âŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
        results.append({
            "URL": url,
            "ì œëª©": "ì˜¤ë¥˜ ë°œìƒ",
            "ë³¸ë¬¸": ""
        })

# ì¢…ë£Œ
driver.quit()

# ê²°ê³¼ ì €ì¥
df_output = pd.DataFrame(results)
df_output.to_csv(output_csv, index=False, encoding='utf-8-sig')

print(f"âœ… ì „ì²´ ì™„ë£Œ. ê²°ê³¼ ì €ì¥: {output_csv}")
```

```python
import pandas as pd
import re

# ê¸°ì¡´ CSV ë¡œë“œ
df = pd.read_csv("output.csv")

def clean_text(text):
    if pd.isna(text):
        return ""

    # ì§‘í•„ìë¶€í„° ëê¹Œì§€ ì œê±° (ì¤„ ë‹¨ìœ„)
    text = re.split(r'\n\s*ì§‘í•„ì\s*:', text)[0]

    # ëª¨ë“  ì¤„ë°”ê¿ˆ ì œê±° (ë¹ˆ ì¤„ í¬í•¨)
    text = text.replace('\n', '').replace('\r', '').strip()

    return text

# 'ë³¸ë¬¸' ì •ì œ
df["ë³¸ë¬¸_ì •ì œ"] = df["ë³¸ë¬¸"].apply(clean_text)

# ê²°ê³¼ ì €ì¥
df.to_csv("cleaned_output.csv", index=False, encoding='utf-8-sig')

print("âœ… ê²°ê³¼ ì €ì¥ ì™„ë£Œ: cleaned_output.csv")
```

```python
import pandas as pd

# íŒŒì¼ ì½ê¸°
df = pd.read_csv("input.csv")

# ë§¤í•‘ ì •ì˜
sortB_map = {
    'ì§„ì‚¬': 'ì§„ì‚¬', 'ìƒì›': 'ìƒì›',
    'ì§„ì‚¬ì‹œ': 'ì§„ì‚¬', 'ìƒì›ì‹œ': 'ìƒì›'
}
sortC_map = {
    'ì‹ë…„': 'ì‹ë…„ì‹œ',
    'ì¦ê´‘': 'ì¦ê´‘ì‹œ',
    'ì‹ë…„ì‹œ': 'ì‹ë…„ì‹œ',
    'ì¦ê´‘ì‹œ': 'ì¦ê´‘ì‹œ',
    'ì¤‘ì‹œ': 'ì¤‘ì‹œ',
    'ë³„ì‹œ': 'ë³„ì‹œ',
    'ì¹œì‹œ': 'ì¹œì‹œ',
    'ì•Œì„±ì‹œ': 'ì•Œì„±ì‹œ',
    'í•´ì£¼ë³„ì‹œ': 'í•´ì£¼ë³„ì‹œ',
    'ì „ì£¼ë³„ì‹œ': 'ì „ì£¼ë³„ì‹œ',
    'ì¶˜ë‹¹ëŒ€ì‹œ': 'ì¶˜ë‹¹ëŒ€ì‹œ'
}
sortD_list = ['ì´ˆì‹œ', 'ë³µì‹œ', 'íšŒì‹œ', 'ì „ì‹œ']
sortE_list = ['ì´ˆì¥', 'ì¤‘ì¥', 'ì¢…ì¥']

# name ì—´ íŒŒì‹± í•¨ìˆ˜
def parse_name(name):
    tokens = str(name).split("_")
    sortB = sortC = sortD = sortE = None

    for token in tokens:
        token = token.strip()
        if sortB is None and token in sortB_map:
            sortB = sortB_map[token]
        elif sortC is None and token in sortC_map:
            sortC = sortC_map[token]
        elif sortD is None and token in sortD_list:
            sortD = token
        elif sortE is None and token in sortE_list:
            sortE = token

    return pd.Series([sortB, sortC, sortD, sortE])

# ì ìš©
df[['sortB', 'sortC', 'sortD', 'sortE']] = df['name'].apply(parse_name)

# ê²°ê³¼ ì €ì¥
df.to_csv("output.csv", index=False)
```
